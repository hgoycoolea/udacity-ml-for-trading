```python
import pandas as pd
```

```python
df = pd.read_csv('data/AAPL.csv', index_col='Date',
                 parse_dates=True, usecols=['Date', 'Adj Close'],
                 na_values=['nan'])
df = df.loc['2020-02-24':'2020-02-28']
```

```python
df.plot()
```

```python
df.fillna(method='ffill').plot()
```
