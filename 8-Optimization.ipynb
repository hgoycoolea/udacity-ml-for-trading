## Finding the minimum of a quadratic function

```python
def f(x):
    return x ** 2 + 4
```

```python
a = np.linspace(-10, 10, 100)
plot(a, f(a))
```

```python
guess = -7
min_result = spo.minimize(f, guess, method='SLSQP', options={'disp': 1})
'x: ' + str(min_result.x), 'y: ' + str(min_result.fun)
```

```python
plot(a, f(a))
plot(min_result.x, min_result.fun, 'rx')
```

## Fitting a linear function

```python
a, b = 3,50
data = np.arange(0, 100) * a + b
data = data + 40 * np.random.random(100)
plot(data, 'rx')
```

```python
def cost(line, data):
    yhat = np.arange(100) * line[0] + line[1]
    return np.sum(np.square(data - yhat))
```

```python
guess = np.random.random(2)
result = spo.minimize(cost, guess, args=(data,), method='SLSQP', options={'disp': True})
```

```python
plot(data, 'rx')
plot(np.arange(100) * result.x[0] + result.x[1])
```

## Fitting a polynomial function

```python
a, b = 3, 4
r = np.arange(-50, 50)
r2 = np.square(r)
data = r2 * a + r * b
data = data + 1000 * np.random.random(100)
plot(data, 'rx')
```

```python
def cost(line, data):
    yhat = r2 * line[0] + r * line[1]
    return np.sum(np.square(data - yhat))
```

```python
guess = np.random.random(2)
result = spo.minimize(cost, guess, args=(data,), method='SLSQP', options={'disp': True})
```

```python
plot(data, 'rx')
r = np.arange(-50, 50)
plot(r2 * result.x[0] + r * result.x[1])
```
